# Step 1- Identifying Missing Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
#Save missing values becuase we might use them later in our analysis. We extract all columns with missing values

#we use list comprehension to achive this...."Read more about list comprehenssion in Python!!"

missing_values = [     feature for feature in df.columns if df[feature].isnull().sum() > 0      ]
```

```python
missing_values
```

```python
#What is the percentage it holds in the missing values; Principal Component Analysis. 
#The ideal thing is to loop through our missing values

for feature in missing_values:
    print(feature, np.round(df[feature].isnull().mean() * 100), '% missing values'                      )
```

```python
# Missing Values Analysis: How the MIssing Values Affects The Price?

for feature in missing_values:
    df[feature]  = np.where(df[feature].isnull(), 1,0)
    df.groupby(feature)['SalePrice'].median().plot.bar()
plt.title('Missing Values Vs SalesPrice')
plt.xlabel('Features')
plt.ylabel('Sales Price')
```
