```python
#Change BloodPressure to Numeric and Seperate the Column into Systolic and Dystolic

# Split the 'Blood Pressure' column into two columns 'Systolic' and 'Dystolic'
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split("/", expand=True)
```

```python
# Convert the 'Systolic' and 'Dystolic' columns to int
df['Systolic'] = df['Systolic'].astype(int)
df['Diastolic'] = df['Diastolic'].astype(int)
```
